```python
import pandas as pd

content = pd.read_csv('Content.csv')
reaction=pd.read_csv('Reactions.csv')
reactiontypes = pd.read_csv('ReactionTypes.csv')
```

We have three datasets which contains the information about the business operations of a social media company. The tasks are:
How many unique categories are there?
How many reactions are there to the most popular category?
What was the month with the most posts?

There was initially 7 datasets, but since we were asked to obtain answers to the questions above, we only needed to specifically choose 3 datasets that would provide us with the answer. These datasets are imported above.

Step 3:
Check for the different types in the "Type" (Or "Content Type", this will be changed later down below) column in both the tables. 

```python
content['Type'].unique()
content['Category'].unique()

```

Code below automatically strips "" for all the strings in the column 'Category'. Hence this column had 41 unique values, but now it has only 29 unique values.

```python
count=0
while count<=999:
   
   content['Category'].replace(content['Category'][count],content['Category'][count].strip('"'),inplace=True)
   count+=1
    
```

```python
content['Category'].unique()
```

```python
content
```

```python
#no need for user id, since we are only checking for popularity of the content
content
```

```python
content['Category'].unique()
```

```python
count=0
while count<=999:
   
   content['Category'].replace(content['Category'][count],content['Category'][count].lower(),inplace=True)
   count+=1
    
```

We make all the contents in the 'Category' column lower case. Hence the unique values in the Category column reduces from 29 to 16.

```python
content['Category'].nunique()
```

```python
del reaction['User ID']
```

```python
reaction
```

```python
content.dropna(inplace=True)
content
```

```python
reactiontypes
reactiontypes.rename(columns = {'Type':'Type_x'}, inplace = True)
```

```python
content
```

Combining the reaction and content dataset using the inner join. 

```python
first_combine = pd.merge(reaction,content,on='Content ID',how='inner')
```

```python
first_combine
```

Now we take the remaining dataset and combine with the ones we already merged.

```python
second_combine =pd.merge(first_combine,reactiontypes,on='Type_x',how='inner')
```

```python
second_combine
```

```python
second_combine
```

Final merged dataset.

```python
second_combine['month'] = pd.DatetimeIndex(second_combine['Datetime']).month
second_combine
```

Using the combined table , we can now find the month which had the maximum viewership. As seen from the result below, the month of May had the maximum viewership.

```python
second_combine['month'].value_counts()
```

The visualization was done in excel using pivot tables. This revealed that the animals category was the most watched category among all other categories. The company had a total of 16 content categories and the month of May had the highest viewership.
